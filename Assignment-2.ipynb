```python
## Bivarient Analysis
```

```python
h1 = pd.read_csv('House_Price_India.csv')
h1.head()
```

```python
x_a='living area'
y_a='lot area'
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=h1, x=x_a, y=y_a)
plt.title(f'Scatter Plot of {x_a} vs. {y_a}')
plt.xlabel(x_a)
plt.ylabel(y_a)
plt.show()
```

```python
categorical_column = 'Built Year '
plt.figure(figsize=(8, 6))
sns.boxplot(data = h1, x='Built Year' , y='living area')
plt.title('Box Plot of living area by Built Year')
plt.xlabel('Built Year')
plt.ylabel('living area')
plt.show()
```

```python
second_categorical_column = 'number of floors '
plt.figure(figsize=(10, 6))
sns.countplot(data=h1, x='Built Year', hue='number of floors')
plt.title(f'Count Plot of {categorical_column} by number of floors')
plt.xlabel('Built Year')
plt.ylabel('Count')
plt.legend(title='number of floors')
plt.show()
```

```python
correlation_coefficient = h1[x_a].corr(h1[y_a])
print(f'Correlation coefficient between {x_a} and {y_a}: {correlation_coefficient}')
```
